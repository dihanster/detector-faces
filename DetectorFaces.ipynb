## Leitura da imagem
Para ler a imagem basta passar o caminho do arquivo de foto a ser analisado. Em seguida, será demonstrado uma visualização da imagem.

```
caminho_imagem = 'drive/My Drive/Imagens/000000177489.jpg'
img = cv2.imread(caminho_imagem )
plt.imshow(img[...,::-1])
plt.show()
```

##Detecção das faces
Para detectar as faces basta utilizarmos o método *detect_face*, passando o objeto imagem. Esse método retorna as coordenadas das faces encontradas e as confianças atribuídas. 
Dessa forma, o próximo passo é percorrer, para cada face detectada, as coordenadas e a partir disso, montar um retângulo (*bouding-box*) ao redor da face detectada.
Feito isso, podemos visualizar a imagem final com as *bouding-box* de detecção.

```
faces, confiancas = cv.detect_face(img)

for face in faces:
    (comeco_x, comeco_y) = face[0], face[1]
    (fim_x, fim_y) = face[2], face[3]
    cv2.rectangle(img, (comeco_x, comeco_y), (fim_x, fim_y), (0, 255, 0), 2)

plt.imshow(img[...,::-1])
plt.show()
```
